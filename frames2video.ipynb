```python
import os
import cv2
from pathlib import Path
DATASET_ROOT = "Bronchoscopy_test"
FPS          = 10
dataset_root = Path(DATASET_ROOT)
clip_dirs = sorted([
    d / "video"
    for seq in dataset_root.iterdir() if seq.is_dir()
    for d in seq.iterdir()
    if d.is_dir() and d.name.startswith("Videos") and (d / "video").exists()
])
print(f"Found {len(clip_dirs)}  clips")
```

```python
success, skipped, failed = [], [], []
for video_dir in clip_dirs:
    output_path = video_dir / "input_video.mp4"
    rel = str(output_path.relative_to(dataset_root))
    if output_path.exists():
        cap = cv2.VideoCapture(str(output_path))
        n_exist = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        cap.release()
        if n_exist > 0:
            print(f"[skip] {rel}  ({n_exist}  frames)")
            skipped.append(rel)
            continue
    imgs = sorted([
        f for f in video_dir.iterdir()
        if f.suffix.lower() in ('.jpg', '.jpeg', '.png')
    ])
    if not imgs:
        print(f"[Warning] no images: {rel}")
        failed.append(rel)
        continue
    first = cv2.imread(str(imgs[0]))
    if first is None:
        print(f"[Error] failed to read first frame: {imgs[0]}")
        failed.append(rel)
        continue
    h, w = first.shape[:2]
    tmp_path = video_dir / "_tmp_video.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    writer = cv2.VideoWriter(str(tmp_path), fourcc, FPS, (w, h))
    if not writer.isOpened():
        print(f"[Error] VideoWriter cannot open: {tmp_path}")
        failed.append(rel)
        continue
    count = 0
    for img_path in imgs:
        img = cv2.imread(str(img_path))
        if img is None:
            print(f"  [dropped frame] {img_path.name}")
            continue
        writer.write(img)
        count += 1
    writer.release()
    if tmp_path.exists() and count > 0:
        if output_path.exists():
            output_path.unlink()
        tmp_path.rename(output_path)
        print(f"OK  {rel:<70}  {count}  frames  {count/FPS:.1f}s  {w}x{h}")
        success.append(rel)
    else:
        print(f"[failed] {rel}")
        failed.append(rel)
print(f"\nDone: {len(success)} generated  |  {len(skipped)} skipped  |  {len(failed)} failed")
```

```python
all_videos = sorted(dataset_root.rglob("input_video.mp4"))
print(f"{'Video path':<70} {'Frames':>6} {'FPS':>5} {'Duration':>7}")
print("-" * 94)
for vp in all_videos:
    cap = cv2.VideoCapture(str(vp))
    n   = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    cap.release()
    dur  = n / fps if fps > 0 else 0
    flag = "!!! " if n == 0 else "    "
    rel  = str(vp.relative_to(dataset_root))
    print(f"{flag}{rel:<70} {n:>6} {fps:>5.1f} {dur:>6.1f}s")
print(f"\nTotal {len(all_videos)} videos")
```
